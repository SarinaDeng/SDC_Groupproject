```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
pd.set_option('display.max_rows', 300) # specifies number of rows to show
pd.options.display.float_format = '{:40,.4f}'.format # specifies default number format to 4 decimal places
plt.style.use('ggplot') # specifies that graphs should use ggplot styling
%matplotlib inline
```

```python
real_other = pd.read_csv('real_other.csv')
```

```python
real_personal = pd.read_csv('real_personal.csv')
```

```python
real_public = pd.read_csv('real_public.csv')
```

```python
plt.rcParams["axes.grid"] = False
f = plt.figure(figsize=(10, 7))
plt.matshow(real_other.corr(), fignum=f.number)
plt.xticks(range(real_other.shape[1]), real_other.columns, fontsize=14, rotation=45)
plt.yticks(range(real_other.shape[1]), real_other.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
```
